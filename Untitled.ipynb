```python
weekdays = ['Mon','Tue','Wed','Thurs','Fri','Sat','Sun']
df_dates = df.dropna(subset = 'Date',axis =0)
df_dates['Date']
df_dates['day'] = df_dates['Date'].apply(lambda x: weekdays[datetime.strptime(x,'%d-%b-%y').weekday()])
```

```python
df_dates.sort_values(by = 'day')
```

### The weekend is from Friday 5:00 p.m. to Sunday 3:00 a.m. . Since for Friday we only have the record of midnight hour none of the Zeroes from Friday are considered. The Zeroes for Saturday and Sunday are included

```python
df_dates.loc[df_dates['day'].isin(['Sat','Sun'])]['Zeroes '].sum()/df_dates['Zeroes '].sum()
```

### As we see around 40.7% of all Zeroes are present on weekends
__Note__ : This percentage is calculated only from the records for which the date is present
